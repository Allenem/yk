```python
# 最小堆栈
# 设计一个支持push,pop,top,并在常量时间内检索最小元素的堆栈.
# getMin()--检索堆栈中的最小元素.

class MinStack:
    def __init__(self):
        self.stack = []
        self.min_stack = []

    def push(self, x):
        self.stack.append(x)
        if not self.min_stack or x <= self.min_stack[-1]:
            self.min_stack.append(x)

    def pop(self):
        if self.stack.pop() == self.min_stack[-1]:
            self.min_stack.pop()

        # if self.stack[-1] == self.min_stack[-1]:
        #     self.min_stack.pop()
        # self.stack.pop()

    def top(self):
        return self.stack[-1]

    def getMin(self):
        return self.min_stack[-1]
    
ministack = MinStack()
ministack.push(-2)
ministack.push(0)
ministack.push(-3)
print(ministack.getMin())
ministack.pop()
print(ministack.top())
print(ministack.getMin())
```

```python
# 一个数组的两个堆栈
# ·描述如何使用单个数组来实现两个堆栈

class ArrayStack:
    def __init__(self, size):
        self.size = size
        self.array = [None] * size
        self.top1 = -1
        self.top2 = size

    def push1(self, x):
        if self.top1 + 1 < self.top2:
            self.top1 += 1
            self.array[self.top1] = x
        else:
            print('stack1 full')
    
    def push2(self, x):
        if self.top1 + 1 < self.top2:
            self.top2 -= 1
            self.array[self.top2] = x
        else:
            print('stack2 full')

    def pop1(self):
        tmp = None
        if self.top1 >= 0:
            tmp = self.array[self.top1]
            self.array[self.top1] = None
            self.top1 -= 1
            return tmp
        else:
            print('stack1 empty')

    def pop2(self):
        tmp = None
        if self.top2 < self.size:
            tmp = self.array[self.top2]
            self.array[self.top2] = None
            self.top2 += 1
            return tmp
        else:
            print('stack2 empty')

    def top11(self):
        if -1 < self.top1 < self.size:
            return self.array[self.top1]
        print('stack1 empty')

    def top22(self):
        if -1 < self.top2 < self.size:
            return self.array[self.top2]
        print('stack2 empty')
    
# size = 10
# array = [None, None, None, None, None, None, None, None, None, None]
# top1 = -1
# top2 = 10

# push1(1)
# array = [1, None, None, None, None, None, None, None, None, None]
# top1 = 0
# top2 = 10

# push2(100)
# array = [1, None, None, None, None, None, None, None, None, 100]
# top1 = 0
# top2 = 9

arrs = ArrayStack(10)
arrs.push1(1)
arrs.push2(100)
print(arrs.array)
print(arrs.pop1())
print(arrs.array)
print(arrs.top1)
print(arrs.top2)
print('--------------------------------------')
print(arrs.top11())
print(arrs.top22())
print(arrs.pop2())
print(arrs.array)
print(arrs.top1)
print(arrs.top2)
print('--------------------------------------')
print(arrs.top11())
print(arrs.top22())
print(arrs.pop1())
print(arrs.pop2())
```

```python
# 一个数组的三个堆栈
# ·描述如何使用单个数组来实现三个堆栈

class Array3Stacks:
    def __init__(self, size):
        self.size = size
        self.array = [None] * size
        self.top1 = -3
        self.top2 = -2
        self.top3 = -1

    def push1(self, x):
        if self.top1 + 3 < self.size:
            self.top1 += 3
            self.array[self.top1] = x
        else:
            print('stack1 full')

    def push2(self, x):
        if self.top2 + 3 < self.size:
            self.top2 += 3
            self.array[self.top2] = x
        else:
            print('stack2 full')

    def push3(self, x):
        if self.top3 + 3 < self.size:
            self.top3 += 3
            self.array[self.top3] = x
        else:
            print('stack3 full')

    def pop1(self):
        tmp = None
        if self.top1 >= 0:
            tmp = self.array[self.top1]
            self.array[self.top1] = None
            self.top1 -= 3
            return tmp
        else:
            print('stack1 empty')

    def pop2(self):
        tmp = None
        if self.top2 >= 1:
            tmp = self.array[self.top2]
            self.array[self.top2] = None
            self.top2 -= 3
            return tmp
        else:
            print('stack2 empty')

    def pop3(self):
        tmp = None
        if self.top3 >= 2:
            tmp = self.array[self.top3]
            self.array[self.top3] = None
            self.top3 -= 3
            return tmp
        else:
            print('stack3 empty')

    def top11(self):
        if -3 < self.top1 < self.size:
            return self.array[self.top1]
        print('stack1 empty')

    def top22(self):
        if -2 < self.top2 < self.size:
            return self.array[self.top2]
        print('stack2 empty')

    def top33(self):
        if -1 < self.top3 < self.size:
            return self.array[self.top3]
        print('stack3 empty')


# arr = [0forstk1, 1forstk2, 2forstk3, 3forstk1, 4forstk2, 5forstk3, 6forstk1, 7forstk2, 8forstk3, 9forstk1]
# stk1     ↑                              ↑                                ↑                          ↑
# stk2                  ↑                              ↑                           ↑                       
# stk3                           ↑                              ↑                            ↑             

arr3s = Array3Stacks(10)
arr3s.push1(0)
arr3s.push2(1)
arr3s.push3(2)
arr3s.push1(3)
arr3s.push2(4)
arr3s.push3(5)

print(arr3s.array)
print(arr3s.pop1())
print(arr3s.array)
print(arr3s.top1)
print(arr3s.top2)
print(arr3s.top3)
print('--------------------------------------')

print(arr3s.top11())
print(arr3s.top22())
print(arr3s.top33())

print(arr3s.pop2())
print(arr3s.array)
print(arr3s.top1)
print(arr3s.top2)
print(arr3s.top3)
```

```python
# 堆栈排序
# ·编写一个程序,按升序排序堆栈,对于堆栈是如何实现的,你不应该做任何假设,下面是编写此程序时应该使用的唯一函数: push|pop|peek|isEmpty

class stack:
    def __init__(self):
        self.stk = []

    def push(self, x):
        self.stk.append(x)

    def pop(self):
        if self.stk:
            return self.stk.pop()
        return None

    def peek(self):
        if self.stk:
            return self.stk[-1]
        return None

    def isEmpty(self):
        return not self.stk
    
def sortStack(stk: stack):
    tmp = stack()
    while not stk.isEmpty():
        cur = stk.pop()
        while not tmp.isEmpty() and tmp.peek() > cur:
            stk.push(tmp.pop())
        tmp.push(cur)
    return tmp


# stk = [3, 2, 1, 4, 5]
# tmp = []

# cur = 3
# tmp = [3]

# cur = 2
# tmp.peek() > cur
# tmp.pop() = 3
# stk.push(3)
# stk = [1, 4, 5, 3]
# tmp = [2]

# cur = 1
# tmp.peek() > cur
# tmp.pop() = 2
# stk.push(2)
# stk = [4, 5, 3, 2]
# tmp = [1]

# cur = 4
# stk = [5, 3, 2]
# tmp = [1, 4]

# cur = 5
# stk = [3, 2]
# tmp = [1, 4, 5]

# cur = 3
# tmp.peek() > cur
# tmp.pop() = 5
# stk.push(5)
# stk = [2, 5]
# tmp.peek() > cur
# tmp.pop() = 4
# stk.push(4)
# stk = [2, 5, 4]
# tmp = [1, 3]

# cur = 2
# tmp.peek() > cur
# tmp.pop() = 3
# stk.push(3)
# stk = [5, 4, 3]
# tmp = [1, 2]

# cur = 5
# stk = [4, 3]
# tmp = [1, 2, 5]

# cur = 4
# tmp.peek() > cur
# tmp.pop() = 5
# stk.push(5)
# stk = [3, 5]
# tmp = [1, 2, 4]

# cur = 3
# tmp.peek() > cur
# tmp.pop() = 4
# stk.push(4)
# stk = [5, 4]
# tmp = [1, 2, 3]

# cur = 5
# stk = [4]
# tmp = [1, 2, 3, 5]

# cur = 4
# tmp.peek() > cur
# tmp.pop() = 5
# stk.push(5)
# stk = [5]
# tmp = [1, 2, 3, 4]

# cur = 5
# stk = []
# tmp = [1, 2, 3, 4, 5]

stk = stack()
stk.push(3)
stk.push(2)
stk.push(1)
stk.push(4)
stk.push(5)

print(stk.stk)
print(sortStack(stk).stk)
```
